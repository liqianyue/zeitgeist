# 加载数据集

```python
#训练集
train_data=pd.read_csv("./train8.3.csv")
#测试集
test_data=pd.read_csv("./test8.3.csv")
#训练集标签
train_label=pd.read_csv("./y_train.csv")
train_label=train_label.iloc[:,1]
```

```python
train_data=train_data.iloc[:,1:].values
train_label=train_label.values.ravel()
print(train_data.shape)
print(train_label.shape)
test_data=test_data.iloc[:,1:].values
print(test_data.shape)
```

```python
model_smote=SMOTE(k_neighbors=20)
X,y=model_smote.fit_resample(train_data,train_label)
X.shape,y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=4)
```

# 模型训练

```python
parameters={
             'objective': 'binary',
             'metric': 'binary_error',
             'verbosity': -1,
             'boosting_type': 'gbdt',
             'lambda_l1': 4.4670221807611144e-08,
             'lambda_l2': 0.20148032727104892,
             'num_leaves': 406,
             'learning_rate': 0.009346454191897936,
             'n_estimators': 4456,
             'feature_fraction': 0.7722750380563909,
             'bagging_fraction': 0.925758056175618,
             'bagging_freq': 7,
             'min_child_samples': 30,
             'max_depth': 44
            }
```

```python
gbm = lgb.LGBMClassifier(**parameters)
```

```python
gbm.fit(X, y,eval_set=[(X,y),(X_test,y_test)],early_stopping_rounds=100)
```

```python
pro_train_LGB_8_3=gbm.predict_proba(train_data)
pro_test_LGB_8_3=gbm.predict_proba(test_data)
pro_train_LGB_8_3=pd.DataFrame(pro_train_LGB_8_3)
pro_test_LGB_8_3=pd.DataFrame(pro_test_LGB_8_3)
pro_train_LGB_8_3.to_csv("./pro_train_lgb_8_3.csv",header=None,index=0)
pro_test_LGB_8_3.to_csv("./pro_test_lgb_8_3.csv",header=None,index=0)
```
