### heatmap（热力图）

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

#原始数据读入
data=pd.read_csv("C:/Users/Lenovo/Desktop/rate_1.csv",names=range(20))
#构建画布对象
f, ax= plt.subplots(figsize = (14, 10))

#需要呈现热力图的数据
corr = data.corr(method='pearson',min_periods=1)
#热力图
#cmap：风格    xticklabels，yticklables横纵坐标     方块内显示数字 annot   如果显示数字格式 fmt='.1f'
sns.heatmap(corr,cmap='RdBu', linewidths = 0.05,ax = ax,xticklabels=range(20),yticklabels=range(20))

# 设置Axes的标题
ax.set_title('Correlation between features')
plt.show()
```

### clustermap（聚类图）

1.method：聚类算法  
    默认为为average算法，可选’single’，’complete’ ,’weighted’,’centroid’,’median’
    
2.metric：簇之间距离选择
    默认为欧式距离euclidean，可选距离： ‘braycurtis’, ‘canberra’, ‘chebyshev’, ‘cityblock’等  
    
3. row_cluster、col_cluster：行列聚类开关 

4. cbar_pos=(left, bottom, width, height)：图例的位置大小 

5. annot 方块内显示数字 

```python
sns.clustermap(corr, method='average', metric='euclidean',
                  figsize=(10, 10), row_cluster=True, col_cluster=True, cmap='RdBu_r',
                cbar_pos=(0.02, 0.8, 0.05, 0.18))
```
